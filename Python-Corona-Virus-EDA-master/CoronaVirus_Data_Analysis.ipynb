```python
df2 = df.groupby(['Country/Region', 'ObservationDate'])[['Country/Region', 'ObservationDate','Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

```python
df3  =df2[df2['Confirmed'] > 50000]
```

```python
countries=df2['Country/Region'].unique()
len(countries)
countries
```

```python
for idx in range(0, len(countries)):
    C=df2[df2['Country/Region']==countries[idx]].reset_index()
    plt.scatter(np.arange(0, len(C)), C['Confirmed'], color='blue', label='Confirmed')
    plt.scatter(np.arange(0, len(C)), C['Recovered'], color='green', label='Recovered')
    plt.scatter(np.arange(0, len(C)), C['Deaths'], color='red', label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since the Record')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
```

```python
df3 = df2.groupby(['ObservationDate'])[['ObservationDate','Confirmed','Deaths','Recovered']].sum().reset_index()
```

```python
C = df3
plt.scatter(np.arange(0,len(C)),C['Confirmed'],color='blue',label='Confirmed')
plt.scatter(np.arange(0,len(C)),C['Recovered'],color='green',label='Recovered')
plt.scatter(np.arange(0,len(C)),C['Deaths'],color='red',label='Deaths')
plt.title('World')
plt.xlabel('Days since the first suspect')
plt.ylabel('Number of cases')
plt.legend()
plt.show()
```
